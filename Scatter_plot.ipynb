```python
#importing bokeh and pandas
import pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show
from bokeh.sampledata.iris import flowers
```

```python
#Define the oputput file
output_file("Scatter_Iris.html")
```

```python
#Create a figure object
f = figure()
```

```python
flowers
```

## How to change the width and hight | Background Color | Transparency

```python
#stylize the plot area
f.plot_width = 1100
f.plot_height = 650
```

## Background Color and Transparency

```python
f.background_fill_color ='orange'
f.background_fill_alpha = 0.5
```

## Border Color

```python
f.border_fill_color ='grey'
```

## Add and Stylize the Title

```python
#Title Text
f.title.text = 'Iris Morphology'
#Title Color
f.title.text_color = 'red'
#Title Size
f.title.text_font_size = "44px"
#Title Font
f.title.text_font ="times"
#Title Alignment
f.title.align ="center"
```

```python
#Adding Glyph
f.circle(x=flowers["petal_length"],y=flowers["petal_width"])
```

```python
show(f)
```

```python
# Use these attributes
dir(f)
```

## Modify Axes Styling

```python
#Style the axex
f.axis.minor_tick_line_color = "blue"
#Major Label Orientation
f.yaxis.major_label_orientation = "vertical"
#To hide the lable
f.xaxis.visible =True #False
#To remove minor ticks
f.axis.minor_tick_line_color = None
```

## Add Axis Lable

```python
f.xaxis.axis_label = "Petal Length"
f.yaxis.axis_label = "Petal Width"
f.axis.axis_label_text_color = "red"
f.axis.major_label_text_color = "blue"
```

# Axes Geometry (Range of the Axes)

```python
from bokeh.models import Range1d
f.x_range = Range1d(start = 0,end =10)
f.y_range = Range1d(start = 0,end =5)
#Add Bound
#f.xaxis.bounds = (2,5)
#Control Ticker Number
f.xaxis[0].ticker.desired_num_ticks =5
```

```python
show(f)
```

# Style the Grid

```python
f.xgrid.grid_line_color = "blue"
f.ygrid.grid_line_color = "red"
```

```python
#Add dashlines
f.grid.grid_line_dash = [5,3]
```

# Style the Tools/ Add or remove the tools

```python
from bokeh.models import PanTool,ResetTool, HoverTool
f.tools=[PanTool(),ResetTool()]
```

```python
# Check all the tools
import bokeh.models
dir(bokeh.models.tools)
```

# Remove Bokeh Logo

```python
f.toolbar.logo =None
```

# Style the Glyphs

```python
f.circle(x=flowers["petal_length"],y=flowers["petal_width"],size =flowers["sepal_width"]*4,fill_alpha = 0.2,color =flowers['color'],legend ='flowers')
```

```python
colormap ={'setosa' :'red','versicolor':'green','virginica':'blue'}
flowers['color']=[colormap[x] for x in flowers['species']]
```

```python
show(f)
```
